Research Question
Is there a significant difference between the population proportions of the participants from Great Britain and the participants from Italy who won Gold medals?

Poplulation:
    All participants from Great Britain and Italy

Parameter of interest: p1-p2
    p1 - Great Britain
    p2 - Italy

Null Hypothesis: p1 - p2 = 0
Alternative Hypthosis: p1 - p2  = 0

```python
GB = data[data.Team == "Great Britain"]
IT = data[data.Team == "Italy"]
```

```python
GB.shape
```

```python
IT.shape
```

```python
n = 2000
GBsample = GB.sample(n)
ITsample = IT.sample(n)
```

```python
gbn = GBsample.Medal.count()
itn = ITsample.Medal.count()
```

```python
print(gbn)
print(itn)
```

```python
perGB = (gbn/n)*100
perIT = (itn/n)*100
```

```python
print(perGB)
print(perIT)
```

```python
n1 = 2000
p1 = 0.1455

n2 = 2000
p2 = 0.157

population1 = np.random.binomial(1, p1, n1)
population2 = np.random.binomial(1, p2, n2)
```
